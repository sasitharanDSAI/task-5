```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.isnull().sum()
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df['deck'] = df['deck'].fillna('Unknown')
```

```python
df
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='pclass', data=df)
plt.title('Passenger Class Count')
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title('Gender Count')
plt.show()
```

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='embark_town', data=df)
plt.title('Embark Town Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='deck', data=df)
plt.title('Deck Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(data=df, x='age', hue='survived', bins=30, kde=True)
plt.title('Age Distribution by Survival (Histogram)')
plt.show()
```

```python
numerical_cols = ['age', 'fare', 'parch', 'sibsp', 'survived']
```

```python
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[numerical_cols], hue='survived')
plt.show()
```

Summary:
- Females had higher survival rates.
- 1st class passengers survived more than 3rd class.
- Younger passengers and very old passengers had different survival rates.
- Deck information was missing for many passengers.
- Embark Town 'Southampton' had most passengers.
